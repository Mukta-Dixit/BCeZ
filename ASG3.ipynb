### (a) Make a DataFrame of all the countries inside the 'Americas' continent. Print the size of the DataFrame and the name list of the countries.

```python
import numpy as np
import pandas as pd
```

```python
asg=pd.read_csv("gapminder.tsv",delimiter="\t")
asg
```

```python
k=asg[asg.continent=="Americas"]
k
```

```python
k.size
```

```python
k1=k.loc[:,["country","gdpPercap"]]
print(k1)

country_america=k1.country.unique()
country_america
```

### (b) Using a loop, calculate the average gdpPercap for all the countries and store them in an array.

```python
avgGDP=[""]*len(country_america)
for i in range (len(country_america)):
    x=k1[k1.country==country_america[i]]
    avgGDP[i]=x.gdpPercap.mean()
avgGDP    
```

```python
avgGDP=pd.DataFrame(avgGDP)
country_america=pd.DataFrame(country_america)
y1=pd.concat((country_america,avgGDP),axis=1)
y1.columns=['country','avg']
y1.set_index('country',inplace=True)
y1
```

### (c) Draw a bar plot of the array in (b) where the x axis shows the names of the countries below the bars and y axis is the average gdpPercap. Axes should be labeled properly and the figure must have a title.

```python
plt=y1.plot(kind="bar",figsize=(10,10),title="ASG3")
plt
```

### (d) Write a program that sorts the values in the array obtained in (b) from the highest to the lowest gdpPercap values. Next, your program should print the three highest gdpPercap values and the three lowest gdpPercap values.

```python
y1=y1.sort_values(by=["avg"],ascending=False)
y1
```
