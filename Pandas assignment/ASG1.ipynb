### (d) Write a program that creates a 2D DataFrame from NumPy arrays. The first array is a column of student ID numbers named 'ID' with values from 1000 to 1049, and the second array is a column of 50 random integer numbers between 60 and 100 named 'Mark'. (hint: use the arange () function from the NumPy package to create the first 'ID' array, and use randint () function from the numpy.random package to create the 'Mark' array. Next, find a way to combine the two arrays together to have a single 2D array, and then convert it to a DataFrame)

```python
data=np.arange(1000,1050).reshape(50,1)
print(data.size)
data1=np.random.randint(60,100,(50,1))
print(data1.size)
combined= np.column_stack((data,data1))
df=pd.DataFrame(combined,columns=['ID','Mark'])
df
```
